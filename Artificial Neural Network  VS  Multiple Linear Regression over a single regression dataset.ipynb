# Regression Model with Artificial Neural Network to Predict value of real estates based on certain Attributes and Comparing the Same with a Multiple Linear Regression Model

```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

## Attribute Information:

####    1. CRIM :            per capita crime rate by town
####    2. ZN :                proportion of residential land zoned for lots over 25,000 sq.ft.
####    3. INDUS :         proportion of non-retail business acres per town
####    4. CHAS :          Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
####    5. NOX :            nitric oxides concentration (parts per 10 million)
####    6. RM :              average number of rooms per dwelling
####    7. AGE  :           proportion of owner-occupied units built prior to 1940
####    8. DIS   :            weighted distances to five Boston employment centres
####    9. RAD  :           index of accessibility to radial highways
####    10. TAX  :          full-value property-tax rate per  (10,000 dollars)
####    11. PTRATIO :   pupil-teacher ratio by town
####    12. B    :            1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
####    13. LSTAT :       % lower status of the population
####    14. MEDV   :      Median value of owner-occupied homes in $1000's 

```python
dataset = pd.read_csv('data.csv')
dataset
```

# Data Preprocessing 

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Converting y into 2d array from 1d array 

```python
y =y.reshape(len(y), 1)
```

## Taking Care of Missing Values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan,strategy='mean')
imputer.fit(X[:, 5:6])
imputer.fit(y)
X[:, 5:6] = imputer.transform(X[:, 5:6])
y=imputer.transform(y)
```

# Splitting the dataset into Training and Test Set 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# Scaling the features for better performance 

```python
from sklearn.preprocessing import StandardScaler
sc_x_1 = StandardScaler()
sc_y = StandardScaler()
sc_x_2 = StandardScaler()
X_train[:, 0:3]= sc_x_1.fit_transform(X_train[:,0:3])
X_train[:, 4:13]= sc_x_2.fit_transform(X_train[:,4:13])
y_train = sc_y.fit_transform(y_train)
```

# Creating Artificial Neural Network Model

```python
ann = tf.keras.models.Sequential()
```

## Input Layer 

```python
ann.add(tf.keras.layers.Dense(500, input_shape=(13,), activation='relu'))
```

## Hidden Layers 

```python
ann.add(tf.keras.layers.Dense(500, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
```

## Output Layer 

```python
ann.add(tf.keras.layers.Dense(1, activation='linear'))
```

## Compiling all the input, hidden and output layers 

```python
ann.compile(optimizer='Adam', loss='mean_squared_error', metrics=['accuracy'])
```

## Training the data with Artificial Neural Network 

```python
ann.fit(X_train, y_train, epochs=1000, verbose=1)
```

## Testing the ANN on test set

```python
X_test[:, 0:3]= sc_x_1.transform(X_test[:,0:3])
X_test[:, 4:13]= sc_x_2.transform(X_test[:,4:13])
```

```python
y_pred = sc_y.inverse_transform(ann.predict(X_test))
```

## Evaluating the ANN model with R-square method

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

# Creating Multiple Linear Regression Model 

## Training the data with Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting and Testing on Test Set

```python
y_pred_lin = sc_y.inverse_transform(regressor.predict(X_test))
```

## Evaluating the Multiple Linear Regression Model with R-square Method

```python
r2_score(y_test,y_pred_lin)
```

# For Artificial Neural Network Model we see that the model evaluation with R-square method gives us an accuracy of approximately 71%



# And for Multiple Linear Regression Model the model evaluation with R-square method gives us an accuracy of approximately 54.64%

# Therefore we can conclude that the ANN model gives us greater advantage of Linear Regression Model in case Regression, as Neural Networks nearly mimics the Human Brain
